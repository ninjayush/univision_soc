```python
# this is what the data from the camera would supposedly look like
raw_detections = [
    {"object": "unauthorized_person", "confidence": 0.92},
    {"object": "stray_cat", "confidence": 0.34},
    {"object": "moving_shadow", "confidence": 0.15},
    {"object": "delivery_truck", "confidence": 0.88}
]

#  filter function
def filter_detections(detections, threshold):
    filtered_results = []
    
    for item in detections:
        # Check if the confidence is greater than or equal to the threshold
        if item["confidence"] >= threshold:
            filtered_results.append(item)
            
    return filtered_results

# We only want detections we are at least 50% sure about- at least here
clean_data = filter_detections(raw_detections, 0.50)
print(clean_data)
```

```python
# Test 1: Standard Case (Should keep high confidence, drop low confidence)
def test_standard_filtering():
    sample = [
        {"object": "person", "confidence": 0.90},
        {"object": "shadow", "confidence": 0.10}
    ]
    result = filter_detections(sample, 0.50)
    
    # We expect only 1 item to survive
    assert len(result) == 1
    assert result[0]["object"] == "person"

# Test 2: Edge Case (What if confidence equals the threshold exactly?)
def test_exact_threshold():
    sample = [{"object": "car", "confidence": 0.50}]
    result = filter_detections(sample, 0.50)
    
    # Based on our '>=' logic, it should keep it
    assert len(result) == 1

# Test 3: Empty Case (What if the camera saw nothing?)
def test_empty_input():
    result = filter_detections([], 0.50)
    assert result == []
```
