```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python

# 1. LOAD: Read the image from your local directory
image = cv2.imread('landscape_original.jpg')

print(image) # as a 3d array
Image(filename='landscape_original.jpg') #direct output via ipython
```

```python

# 2. INSPECT: Print the underlying numpy array shape
# Output will look like (Height, Width, Channels) -> e.g., (1080, 1920, 3)
print(f"Original Image Shape: {image.shape}")
print(f"Top-Left Pixel BGR Values: {image[0, 0]}") 
```

```python

# 3. RESIZE: Shrink the image to half for faster AI processing
# Note: OpenCV takes dimensions as (Width, Height)
shape=tuple((np.array(image.shape)/2)[0:2].astype(int)[::-1]) #reversed as cv2.resize takes W,H
print(shape)
resized_image = cv2.resize(image, shape)
print(resized_image, resized_image.shape) #check

# OpenCV loads images as BGR, so convert to RGB for correct colors
# plt.imshow(resized_image)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # hide axes
plt.title('Resized Image')
plt.show()
```

```python

# 4. GRAYSCALE: Convert 3 channels (BGR) down to 1 channel
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(f"Grayscale Image Shape: {gray_image.shape}") # Notice the '3' is gone!

plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()

cv2.imwrite('landscape_grayscale.jpg', gray_image)
```

```python

# 5. CROP: Slice the numpy array [Start_Y:End_Y, Start_X:End_X]
# This grabs a 200x200 square out of the top-left corner
cropped_image = gray_image[0:200, 0:200]

plt.imshow(cropped_image, cmap='gray')
plt.axis('off')
plt.show()
```
